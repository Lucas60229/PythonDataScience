```
import matplotlib.pyplot as plt
import numpy as np
```

```
x = np.arange(0, 3 * np.pi, 0.1)
y_cos = np.cos(x)
plt.plot(x,y_cos)
# plt.show
plt.savefig("filename.png",dpi=300,format="png")
```

```
scatter_X = np.random.normal(0, 1, 1024)
scatter_Y = np.random.normal(0, 1, 1024)
arcTan = np.arctan2(scatter_Y, scatter_X)
plt.scatter(scatter_X, scatter_Y, c=arcTan)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()
```
